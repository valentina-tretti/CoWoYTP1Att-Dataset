```python
# Defining patterns
re_spaces = re.compile("([\"\''().,;:/_¿?¡!—\-])") # Add spaces around punctuation mark
re_punc = re.compile(r'([.,!?¿¡:;()])')
re_mult_space = re.compile(r"  *") # Changes multiple spaces into one space

# Emoji removal based on  https://gist.github.com/slowkow/7a7f61f495e3dbb7e3d767f97bd7304b
emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                               u"\U00002500-\U00002BEF"  # chinese char
                               u"\U00002702-\U000027B0"
                               u"\U00002702-\U000027B0"
                               u"\U000024C2-\U0001F251"
                               u"\U0001f926-\U0001f937"
                               u"\U00010000-\U0010ffff"
                               u"\u2640-\u2642"
                               u"\u2600-\u2B55"
                               u"\u200d"
                               u"\u23cf"
                               u"\u23e9"
                               u"\u231a"
                               u"\ufe0f"  # dingbats
                               u"\u3030"
                               "]+", flags=re.UNICODE)

# Function to replace numbers with words in Spanish
def replace_numbers_with_words(text):
    return re.sub(r'\d+',
                  lambda match: num2words(int(match.group()), lang='es'),
                  text)



def clean_text(text):
    # Based on https://github.com/juli-amezquita/Topic-Modeling-in-Spanish-Amazon-Reviews/blob/main/Topic_Modeling_en_Espa%C3%B1ol_Amazon_Reviews.ipynb
    text = text.lower()
    text = re.sub(r'\*\*', '', text)  # Remove **
    text = re.sub(r'[\"\']', '', text)  # Remove quotation marks
    text = emoji_pattern.sub('', text)  # Remove emojis

    # Replace unwanted tokens with space
    tokens_to_remove = [r'xdxdxd', r':v', r'xd', r'uwu', r'ᐛ']
    pattern = r'\b(?:' + '|'.join(tokens_to_remove) + r')\b'
    text = re.sub(pattern, ' ', text)


    # Removing XD, XXD, xD, etc.
    text = re.sub(r'\b[x]{1,2}d\b', '', text, flags=re.IGNORECASE)


    # Removing punctuation except @
    punct_to_remove = string.punctuation.replace('@', '') + '¡¿°€ª”:'
    text = re.sub(r'[%s]' % re.escape(punct_to_remove), '', text)

    text = re.sub(r"(.)\1{2,}", r"\1", text)  # Normalizing repeated characters
    text = replace_numbers_with_words(text)  # Converting numbers to words
    text = re_mult_space.sub(' ', text)  # Normalizing whitespace
    return text.strip()
```

```python
# Calling clean function
data["comment_cleaned_w_stopWords"] = data["comment"].apply(clean_text)
```
